```python
import os
import tensorflow as tf
from matplotlib import pyplot as plt
from tensorflow.keras.applications.resnet import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
path = '../../datasets/rawdata'
target_path = '../../datasets/mahasiswa_dataset_preprocessed'
```

```python
for folder in os.listdir(path):
   # print(folder)
   folder_path = os.path.join(path, folder)
   # folder_name = folder.split('_')[1]
   target_folder = os.path.join(target_path, folder)
   
   if not os.path.exists(target_folder):
      os.makedirs(target_folder)
      
   for image in os.listdir(folder_path):
      image_path = os.path.join(folder_path, image)
      img = load_img(image_path)
      img_array = img_to_array(img)
      
      preprocessed_image = preprocess_input(img_array)
      tf.keras.utils.save_img(os.path.join(target_folder, image), preprocessed_image)
      
```
